```python
import json
import pickle
import joblib
import pandas as pd
import numpy as np
```

```python
# load file

# Modelling
with open('model.pkl', 'rb') as file_1:
    model = pickle.load(file_1)


# Pre-processing

with open('pipeline.pkl', 'rb') as file_2:
    num_pipeline = pickle.load(file_2)
    
    
# List Numeric & Category
with open('num_cols.txt', 'r') as file_3:
    num_cols = json.load(file_3)

with open('cat_cols.txt', 'r') as file_4:
    cat_cols = json.load (file_4)
    
```

# 10. Model Inferences

```python
df_inferences = pd.read_csv('ParisHousing.csv')
```

```python
# set numpy option
np.set_printoptions(suppress=True)

# set view in pandas
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df_inferences = df_inferences.head(10)
```

```python
df_prediksi = df_inferences.head(10)
```

```python
df_clean = df_inferences.drop(['price','cityCode' ,'made'], axis=1)
```

```python
num_cols
```

```python
df_clean[num_cols]
```

```python
df_inferences=num_pipeline.transform(df_clean[num_cols])
```

```python
df_inferences = pd.DataFrame(df_inferences , columns=num_cols)
```

```python
df_final = pd.concat([df_inferences,df_clean[cat_cols]], axis=1)
```

```python
df_final
```

```python
df_final =df_final.loc[:,['squareMeters','hasYard','hasPool','isNewBuilt','hasStormProtector','hasStorageRoom']]
```

```python
y_pred = model.predict(df_final)
```

```python
y_pred_inf = pd.DataFrame(y_pred, columns=['prediction'])
```

```python
y_pred_inf
```

```python
df_inf_prediksi = pd.concat([df_prediksi,y_pred_inf] , axis=1)
```

```python
df_inf_prediksi
```

```python
df_inf_prediksi['difference'] = df_inf_prediksi['price'] - df_inf_prediksi['prediction']
```

```python
df_inf_prediksi[['price','prediction','difference']]
```

The difference is considered small from the original value.
